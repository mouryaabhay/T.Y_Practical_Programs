```python
# example
from pulp import *

model = LpProblem(sense = LpMaximize)

x = LpVariable(name="x", lowBound=0)
y = LpVariable(name="y", lowBound=0)
z = LpVariable(name="z", lowBound=0)

model += (x+2*y+z<=430)
model += (3*x+2*z<=460)
model += (x+4*y<=120)

model += (3*x+2*y+5*z)

print("\nModel:\n", model)

print("\nSolution Status: ", model.solve())

print("\nOptimal value for objective function: ", model.objective.value())

print("\nX value: ", x.value())
print("Y value: ", y.value())
print("Z value: ", z.value())
```
