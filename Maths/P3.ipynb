```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#example
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(x,y)
Z = np.cos(X + Y)

fig = plt.figure(figsize=(9,9))
    
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='blue',lw=0.7,rstride=2, cstride=2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('f(x,y)')
plt.show()
```

```python
# SET 1.1
x = np.linspace(0, 2*np.pi, 100)
y = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(x,y)
Z = np.exp(X**2+Y**2)

fig = plt.figure(figsize=(9,9))
    
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='blue',lw=0.7,rstride=2, cstride=2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$f(x,y) = e^{x^{2} + y^{2}}$')
plt.show()
```

```python
#SET 2.1
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x,y)
Z = X*np.exp(Y)

fig = plt.figure(figsize=(9,9))
    
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$f(x,y) = xe^{y}$')
plt.show()
```

```python
#SET 2.2
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x,y)
Z = X ** 3 + Y ** 3 + X ** 2 + Y ** 2 + X + Y + 5

fig = plt.figure(figsize=(9,9))
    
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$f(x,y) = x^{3}+y^{3}+x^{2}+y^{2}+x+y+5$')
plt.show()
```
