```python
from sympy import *
from math import pi

x, y = symbols('x, y')
L = Line(Point(0,0),Point(0,1))
l1 = L.rotate(pi/6)
print("Rotation:", l1)
print("Equation:", l1.equation())
```

```python
s = Segment((1,0),(2,-1))
s.rotate(pi)
```

```python
r = Ray((0,0),(4,4))
print("Anti Clockwise: ", r.rotate(pi/2))
print("Clockwise: ", r.rotate(-pi/2))
```

```python
l1 = Line(4*x+3*y-5)
l2 = Line(x+y)
l3 = l1.reflect(l2)
print("Reflection: ", l3)
print("Equation: ", l3.equation())
```

```python
s = Segment((2,3),(4,6))
l = Line(7*x+6*y-3)
rs = s.reflect(l)
print("Reflection: ", rs)
```

```python
r = Ray((0,0),(2,4))
l = Line(x-2*y-3)
rs = r.reflect(l)
print("Reflection: ", rs)
```

```python
A = Point(2,1)
B = Point(4,-1)

T = Matrix([[1,2,0], [2,1,0],[0,0,1]])

l = Line(A.transform(T), B.transform(T))
l.equation()
```

```python
A = Point(1,1)
B = Point(-4,-1)

T = Matrix([[1,-2,0], [-2,1,0],[0,0,1]])

s = Segment(A.transform(T), B.transform(T))

print(s)
```

```python
A = Point(2, 5)
B = Point(4, -13)

T = Matrix([[2,3,0], [4,1,0],[0,0,1]])

A1 = A.transform(T)
B1 = B.transform(T)

S = Segment(A1, B1)
print("Midpoint: ", S.midpoint)
print("Slope: ", S.slope)
```

```python
L = Line(2*x+y-3)

P = L.points[0]
Q = L.points[1]

T = Matrix([[1,-3,0], [2,1,0],[0,0,1]])

l = Line(P.transform(T), Q.transform(T))
l.equation()
```
